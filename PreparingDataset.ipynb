```python
import os
```

```python
def categorize_position(x_center: float) -> str:
    '''The func takes in the center of element as input and return its current position.'''
    if x_center < 0.33:
        return 'left'
    elif x_center > 0.66:
        return 'right'
    else:
        return 'front'
```

```python
def analyze_yolo_file(file_path: str)-> str:
    with open(file_path, 'r') as f:
        line = f.readline()
    
    parts = line.strip().split()
    class_id, x_center, y_center, width, height = map(float, parts)
    result = categorize_position(x_center)
    return result
```

```python
import shutil
```

```python
def process_dataset(split):
    labels_dir = os.path.join("labels", split)
    images_dir = os.path.join("images", split)

    for filename in os.listdir(labels_dir):
        label_path = os.path.join(labels_dir, filename)
        position = analyze_yolo_file(label_path)

        # Create subfolder if doesn't exist
        dest_dir = os.path.join(images_dir, position)
        os.makedirs(dest_dir, exist_ok=True)

        # Corresponding image file
        image_name = filename.replace(".txt", ".jpg")
        image_path = os.path.join(images_dir, image_name)

        if os.path.exists(image_path):
            dest_path = os.path.join(dest_dir, image_name)
            shutil.copy2(image_path, dest_path)
        else:
            print(f"Image not found for: {filename}")
```

```python
process_dataset("train")
```

```python
process_dataset("test")
```
